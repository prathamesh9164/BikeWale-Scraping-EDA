```python
## https://www.bikewale.com/best-scooters-in-india/?sortOrder1=desc&sortField1=modelPopularity&pageNumber=2
```

```python
model_name = []
Engine_size = []
Mileage = []
avg_Rating = []
No_of_Rating = []
price = []

for i in range(2,30):
    urls = "https://www.bikewale.com/best-scooters-in-india/?sortOrder1=desc&sortField1=modelPopularity&pageNumber=" + str(i)
    print(urls)
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
    }
    page = requests.get(urls, headers=headers) #Here instead of url we need to keep urls 
    print(page.status_code)
    
    soup = BeautifulSoup(page.text)

    ## Creating Model Name List

    M = soup.find_all("span",class_="o-jJ o-j3 o-js o-ez")

    for i in M:
        model_name.append(i.text)

    ## Creating Engine Size List

    E = soup.find_all("span",class_="o-iZ o-jf")
    Br = []
    for i in E:
        text = i.text.replace('|','').strip()
        Br.append(text)

    for i in Br:
        if i.strip().endswith('cc') or i.strip().endswith('kWh'):
            Engine_size.append(i)

    ## Creating Mileage List

    M = soup.find_all("span",class_="o-iZ o-jf")
    mil = []
    for i in M:
        text = i.text.replace('|','').strip()
        mil.append(text)

    for i in mil:
        if i.strip().endswith('km') or i.strip().endswith('kmpl') or i.strip().endswith('kml'):
            Mileage.append(i)

    ## Creating Average Rating List

    A_R = soup.find_all("div",class_="o-f7 o-o")
    for i in A_R:
        b = re.findall(r"(\d\.\d)\/",i.text)
        if b:
            avg_Rating.append(b[0])
    
    ## Creating No of Rating List

    N_R = soup.find_all("span",class_="o-jL o-f7 o-j0 o-jc o-fE")
    for i in N_R:
        No_of_Rating.append(i.text.replace('Ratings','').strip())

    ## Creating Price List

    P = soup.find_all("div",class_="o-f7 o-o")
    for i in P:
        price.append(i.text.split("₹")[1].strip().split()[0].replace('Avg.',''))
```

```python
print(len(model_name),len(Engine_size),len(Mileage),len(avg_Rating),len(No_of_Rating),len(price))
```

```python
# Find the maximum length among all lists to avoid losing data
target_length = max(len(model_name), len(Engine_size), len(Mileage), len(avg_Rating), len(No_of_Rating), len(price))

# Helper function to adjust list length in-place
def adjust_length(lst, target_len):
    if len(lst) < target_len:
        lst.extend([np.nan] * (target_len - len(lst)))
    elif len(lst) > target_len:
        del lst[target_len:]

# Adjust all lists to the same target length
adjust_length(model_name, target_length)
adjust_length(Engine_size, target_length)
adjust_length(Mileage, target_length)
adjust_length(avg_Rating, target_length)
adjust_length(No_of_Rating, target_length)
adjust_length(price, target_length)
```

```python
data = pd.DataFrame({
    "Model Name": model_name,
    "Engine Size": Engine_size,
    "Mileage": Mileage,
    "Average Rating": avg_Rating,
    "Number of Ratings": No_of_Rating,
    "Price": price
})
```
